```python
import json
import pandas as pd
import torch

from torch.utils.data import DataLoader

from main import (
    get_model,
    evaluate_metrics
)

from data_utils import (
    Dataset_ASVspoof2019_devNeval,
    genSpoof_list
)

# =====================================
# SETTINGS
# =====================================

DEVICE = "cuda"

CONFIG_PATH = "/content/aasist/config/AASIST.conf"

MODELS = {
    "seed42":
    "/content/drive/MyDrive/AudioDeepfakesCodes/Models/LA_AASIST_ep25_bs24_seed42/weights/best.pth",

    "seed1234":
    "/content/drive/MyDrive/AudioDeepfakesCodes/Models/LA_AASIST_ep25_bs24_seed1234/weights/best.pth",

    "seed2025":
    "/content/drive/MyDrive/AudioDeepfakesCodes/Models/LA_AASIST_ep25_bs24_seed2025/weights/best.pth",
}

DATASETS = {
    "default":
    "/content/drive/MyDrive/AudioDeepfakesCodes/Data/test_16k/default_vs_real.csv",

    "vc_david":
    "/content/drive/MyDrive/AudioDeepfakesCodes/Data/test_16k/vc_david_vs_real.csv",

    "vc_jennifer":
    "/content/drive/MyDrive/AudioDeepfakesCodes/Data/test_16k/vc_jennifer_vs_real.csv",

    "vc_morgan":
    "/content/drive/MyDrive/AudioDeepfakesCodes/Data/test_16k/vc_morgan_vs_real.csv",
}

# =====================================
# LOAD CONFIG
# =====================================

with open(CONFIG_PATH, "r") as f:
    config = json.load(f)

results = []

# =====================================
# EVALUATION LOOP
# =====================================

for model_name, model_path in MODELS.items():

    print("\n" + "=" * 80)
    print("Loading:", model_name)

    model = get_model(
        config["model_config"],
        DEVICE
    )

    state_dict = torch.load(
        model_path,
        map_location=DEVICE
    )

    model.load_state_dict(state_dict)

    model.eval()

    for dataset_name, csv_file in DATASETS.items():

        print(
            f"\nEvaluating {model_name} on {dataset_name}"
        )

        labels, file_list = genSpoof_list(csv_file)

        dataset = Dataset_ASVspoof2019_devNeval(
            list_IDs=file_list,
            labels=labels
        )

        loader = DataLoader(
            dataset,
            batch_size=24,
            shuffle=False,
            drop_last=False,
            pin_memory=True
        )

        acc, prec, rec, f1, eer = evaluate_metrics(
            loader,
            model,
            DEVICE
        )

        results.append({
            "Model": model_name,
            "Dataset": dataset_name,
            "Accuracy": acc,
            "Precision": prec,
            "Recall": rec,
            "F1": f1,
            "EER": eer
        })

        print(
            f"Acc={acc:.4f} "
            f"Prec={prec:.4f} "
            f"Recall={rec:.4f} "
            f"F1={f1:.4f} "
            f"EER={eer:.4f}"
        )

# =====================================
# SAVE RESULTS
# =====================================

results_df = pd.DataFrame(results)

save_path = (
    "/content/drive/MyDrive/"
    "AudioDeepfakesCodes/Models/"
    "robustness_results.csv"
)

results_df.to_csv(
    save_path,
    index=False
)

print("\n")
print(results_df)

print("\nSaved to:")
print(save_path)
```
